```python
##Funciones de ayuda para calcular la inversa complementaria que usaremos a lo largo de este notebook.
def inv(secuencia):
    return secuencia[::-1]

def comp(secuencia):
    ret_str = ''
    secuencia = secuencia.upper()
    for c in secuencia:
        if c == 'A':
            ret_str += 'T'
        elif c == 'C':
            ret_str += 'G'
        elif c == 'T':
            ret_str += 'A'
        elif c == 'G':
            ret_str += 'C'
        else:
            ret_str += c
     
    return ret_str
            
def inv_comp(secuencia):
    return comp(inv(secuencia))
```

```python
def kmeros_frecuentes(secuencia, k):
    freq = {}
    n = len(secuencia)
    for i in range(n-k+1):
        kmero = secuencia[i:i+k]
        if kmero in freq:
            freq[kmero] += 1
        else:
            freq[kmero] = 1
    return freq
```

```python
def top_kmeros(secuencia, k):
    kmeros = []
    freqs = kmeros_frecuentes(secuencia, k)
    m = max(freqs.values())
    for key in freqs:
        if freqs[key] == m:
            kmeros.append(key)
        # add each key to words whose corresponding frequency value is equal to m
    return kmeros, m
```

Para modificar nuestro m칠todo de contar patrones, primero tendremos que permitir hacer comparaciones "difusas" basadas en el n칰mero de car치cteres diferentes entre dos cadenas. Si este n칰mero est치 por debajo de un l칤mite dado, las cadenas ser치n consideradas iguales :)

__Ejercicio__: Implementa una funci칩n `distancia(kmero_1, kmero_2)` que devuelva el n칰mero de caracteres diferentes que hay entre dos k-meros dados. Ahora, genera otra funci칩n `busca_kmero_aprox(secuencia, kmero, d)` que devuelva cu치ntas veces aparece un k-mero (y otros k-meros parecidos con distancia <= d) en una secuencia dada (considera tambi칠n la inversa complementaria del k-mero).


```python
def distancia(kmero_1, kmero_2):
    diferencias = 0
    for a, b in zip(kmero_1, kmero_2):
        if a != b: 
            diferencias+=1
    
    return diferencias
        
```

```python
def cuenta_kmero_aprox(secuencia, kmero, d):
    cuenta = 0
    for i in range(len(secuencia) - len(kmero) + 1):
        if distancia(secuencia[i:i+len(kmero)], kmero) <= d:
            cuenta += 1
        if distancia(secuencia[i:i+len(kmero)], inv_comp(kmero)) <= d:
            cuenta += 1
    return cuenta
```

Como hemos visto, en esta nueva modalidad de b칰squeda difusa, un k-mero puede no estar presente en la cadena y, a pesar de ello, ser el m치s frecuente. Por ello, tendremos que probar con todos los $4^{k}$ k-meros posibles. 

Vamos a implementarlo en c칩digo:

```python
def kmeros_frecuentes_aprox(secuencia, k, d):
    freq = {}
    n = len(secuencia)
    todos_kmeros = list(["".join(kmero) for kmero in itertools.product("ACGT", repeat=k)])
    for kmero in todos_kmeros:
        frec = cuenta_kmero_aprox(secuencia, kmero, d)
        freq[kmero] = frec
    return freq
```

쯈u칠 ocurre aqu칤? Parece que hay bastantes k-meros que aparecen 4 veces en la regi칩n que hemos identificado como el origen de replicaci칩n...쮺u치l de ellos podr칤a ser la caja DnaA?

Vamos a intentar agrupar los resultados considerando que:
1. La caja puede aparecer como una cadena o su inversa complementaria.
2. La caja puede contener mutaciones de, como mucho en este caso, 1 elemento. Por lo tanto, a todos los efectos, las cadenas que difieran en 1 caracter de una cadena cadena (o de su inversa complementaria) ir치n al mismo grupo que dicha cadena.

```python
def agrupa_resultados(top_kmeros):
    grupos = []
    while len(top_kmeros) > 0:
        kmero = top_kmeros.pop()
        grupo_temp = [kmero]
        for i in range(len(top_kmeros)):
            if distancia(kmero, top_kmeros[i]) <= 1 or kmero == inv_comp(top_kmeros[i]) or distancia(inv_comp(kmero), top_kmeros[i]) <= 1:
                grupo_temp.append(top_kmeros[i])

        top_kmeros = [kmero for kmero in top_kmeros if kmero not in grupo_temp]
        grupos.append(sorted(grupo_temp)) # ordenamos cada grupo lexicogr치ficamente para ver sus diferencias.
    
    return sorted(grupos, key=lambda kmeros: kmeros[0]) # mantenemos el orden lexicogr치fico en la lista de grupos
```

### Extra - Optimizando y acelerando nuestros algoritmos 游끭游끠

Como hemos visto, la funci칩n `cuenta_kmero_aprox` tardaba bastante en ejecutarse, ya que un requisito era generar previamente los $4^{k}$ k-meros posibles. Una pregunta que qued칩 en el aire era c칩mo pod칤amos acelerar este algoritmo: vamos a verlo. 

#### Optimizando
La optimizaci칩n en este caso la vamos a conseguir considerando s칩lo el grupo de k-meros que tengan una distancia `d` o inferior a cualquier k-mero que aparezca en la secuencia. A estos grupos, los denominaremos `d-vecindarios`.

Por ejemplo, el 1-vecindario (vecinos inmediatos) del 3-mero `ACG` es: `{ACG, ACG CCG GCG TCG AAG AGG ATG ACA ACC ACT}` (n칩tese que se incluye el "vecino" originario del vecindario). 

Como ves, el tama침o del d-vecindario de un k-mero depender치 del n칰mero de s칤mbolos del alfabeto y la longitud del k-mero.

__Pregunta: 쮺u치ntos vecinos tendr치 el vecindario de un k-mero (para el alfabeto del ADN)?__


Vamos a implementar la funci칩n `vecinos_directos(kmero, alfabeto)`, que nos devolver치 el conjunto de vecinos directos para un k-mero y un alfabeto:

```python
def vecinos_directos(kmero, alfabeto):
    vecindario = set([kmero])
    elementos_kmero = list(kmero)
    for i, b in enumerate(elementos_kmero):
        for letra in alfabeto:
            if letra == b:
                continue
            vecino = elementos_kmero[:]
            vecino[i] = letra
            vecindario.add("".join(vecino))
    return vecindario
```

### d-vecindarios
La funci칩n que hemos implementado genera el 1_vecindario (vecinos directos) para un k-mero, 춰pero recuerda que la b칰squeda difusa puede permitir m치s modificaciones!

__Pregunta: Piensa c칩mo generar칤as el d-vecindario de un k-mero dado.__

```python
def vecindario(kmero, alfabeto, d):
    vecindario = set([kmero])
    for i in range(d):
        i_vecinos = set()
        for vecino in vecindario:
            i_vecinos = i_vecinos | vecinos_directos(vecino, alfabeto)
        vecindario = vecindario | i_vecinos
    return vecindario 
```

### S칩lo los vecinos
Ahora que ya sabemos encontrar el d-vecindario de un k-mero, podemos modificar nuestro algoritmo para generar mapas de frecuencias siguiendo nuestro razonamiento inicial. 

__Pregunta: Modifica la funci칩n kmeros_frecuentes_aprox para que considere s칩lo los k-meros que sean miembros de alg칰n vecindario de todos los k-meros que aparezcan en la secuencia. Mide el tiempo que tarda en ejecutarse con la magia `%time`. 쮼s esta diferencia la que esperabas?__

```python
def kmeros_frecuentes_aprox_vecindarios(secuencia, k, d, alfabeto):
    freq = {}
    cercanos = []
    
    for i in range(len(secuencia) - k + 1):
        vecindario_actual = vecindario(secuencia[i:i+k], alfabeto, d)
        for vecino in vecindario_actual:
            cercanos.append(vecino)
    
    cercanos = list(set(cercanos))
    for cercano in cercanos:
        freq[cercano] = cuenta_kmero_aprox(secuencia, cercano, d)
    
    return freq
```
